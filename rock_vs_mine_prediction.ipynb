```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
sonar_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Machine Learning Projects/Datasets/sonar_data.csv', header=None)
```

```python
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1, stratify=Y, random_state=1)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_pred = model.predict(X_train)
accuracy_train = accuracy_score(X_train_pred, Y_train)
```

```python
print('Train accuracy:', accuracy_train)
```

```python
X_test_pred = model.predict(X_test)
accuracy_test = accuracy_score(X_test_pred, Y_test)
```

```python
print('Test accuracy:', accuracy_test)
```
